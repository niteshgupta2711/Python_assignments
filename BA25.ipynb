```python
def cou(l):
    d={}
    for i in l:
        if i not in d:
            d[i]=1
        else:
            d[i]+=1
    return d
```

```python
cou([3,4,3])
```

```python
def equal(a,b,c):
    l=cou([a,b,c])
    if max(l.values())==1:
        return 0
    else: return max(l.values())
    
```

```python
equal(3,4,3)
```

```python
def tup(dict):
    for k,v in dict.items():
        yield (k,v)
```

```python
def dict_to_list(dict):
    return sorted([x for x in tup(dict)])
    
```

```python
dict_to_list({'d':1,'b':2,'c':3})
```

```python
def mapp(list):
    list1=[x.lower() for x in list]
    return {i : i.upper() for i in list1}
```

```python
def vow_replace(str1,str2):
    l2=[x for x in str1]
    def vow(char):
        if char in 'aeiou':
            return True
    for i,v in enumerate(str1):
        if vow(v):
            l2[i]=str2
    return ''.join(l2)
            
```

```python
vow_replace('apples and bananas','u')
```
